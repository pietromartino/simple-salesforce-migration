Utils

```python
class SimpleMigration:
    
    def __init__(self, source, target):
        self.source = source
        self.target = target
    
    def launch(self, obj):
        self.extract(obj)
        self.migrate(obj)
        return
    
    def migrate(self, obj):
        self.migration_df = self._prepare_source_df_for_migration(obj)
        to_migrate = self.migration_df.to_dict(orient='records')
        self.migration_result = getattr(self.target.bulk, obj).insert(to_migrate)
        return self.migration_result

    def _prepare_source_df_for_migration(self, obj):
        all_cols = get_all_field_names(getattr(self.source, obj))
        keep_cols = remove_system_fields(all_cols)
        keep_cols = remove_insert_fields(keep_cols)
        return self.source_df[keep_cols].replace({np.nan: None}) # needed to avoid errors when serializing in json 
    
    def extract(self, obj):
        qe_source = QueryEngine(self.source)
        self.source_df = pd.DataFrame(qe_source.select_all(obj)['records'])
        return self.source_df


class QueryEngine:
    
    def __init__(self, instance):
        self.sf = instance
        
    def select_all(self, obj_name):
        field_names = get_all_field_names_queryformat(getattr(self.sf, obj_name))
        select_all_query = f'SELECT {field_names} FROM {obj_name}'
        return self.sf.query_all(select_all_query)

    
def get_all_field_names_queryformat(obj):
    fields = get_all_field_names(obj)
    return ','.join(fields)

def get_all_field_names(obj):
    desc = obj.describe()
    field_names = [field['name'] for field in desc['fields']]
    return field_names

def remove_system_fields(field_list):    
    __system_fields = [
        'IsDeleted', 
        'CreatedBy', 
        'CreatedById', 
        'CreatedDate', 
        'LastModifiedBy', 
        'LastModifiedById', 
        'LastModifiedDate',
        'LastViewedDate',
        'SystemModstamp',
        'LastActivityDate',
        'LastReferencedDate',
        'LastCURequestDate',
        'PhotoUrl',
        'JigsawCompanyId',
        'MasterRecordId',
        'BillingAddress',
        'MailingAddress',
        'ShippingAddress',
        
    ]
    return list(set(field_list)-set(__system_fields))

def remove_insert_fields(field_list):
    __insert_fields = [
        'Id', 
        'OwnerId',
        'Name',
        'LastCUUpdateDate',
        'IsEmailBounced',
        'LastCURequestDate',
        'OtherAddress',
        'JigsawContactId'
    ]
    return remove_system_fields(list(set(field_list)-set(__insert_fields)))
    
def remove_custom_fields(field_list):
    return [f for f in field_list if f[-3:] != '__c']
```
